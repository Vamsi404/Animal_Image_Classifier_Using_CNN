```python
import os
import shutil

# Define the path to the val_images directory
val_images_path = 'val_images'

# Create the directory if it doesn't exist
if not os.path.exists(val_images_path):
    os.makedirs(val_images_path)

# Function to extract the class name from the filename
def extract_class_name(filename):
    if '.' in filename and '-' in filename:
        # Find the position of the first occurrence of each delimiter
        dot_position = filename.find('.')
        hyphen_position = filename.find('-')
        # Use the delimiter that appears first
        if dot_position < hyphen_position:
            return filename.split('.')[0]
        else:
            return filename.split('-')[0]
    elif '.' in filename:
        return filename.split('.')[0]
    elif '-' in filename:
        return filename.split('-')[0]
    else:
        return None

# Iterate over all files in the val_images directory
for filename in os.listdir(val_images_path):
    # Skip if it is a directory
    if os.path.isdir(os.path.join(val_images_path, filename)):
        continue

    # Extract the class name from the filename
    class_name = extract_class_name(filename)

    # Continue to next file if class name is not found
    if class_name is None:
        continue

    # Define the path to the class folder
    class_folder_path = os.path.join(val_images_path, class_name)

    # Create the class folder if it doesn't exist
    if not os.path.exists(class_folder_path):
        os.makedirs(class_folder_path)

    # Define the source and destination file paths
    src_file_path = os.path.join(val_images_path, filename)
    dest_file_path = os.path.join(class_folder_path, filename)

    # Move the file to the class folder
    shutil.move(src_file_path, dest_file_path)

print("Files have been organized into respective class folders.")
```

```python
from keras.layers import *
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(MaxPool2D(2,2) )
model.add(Conv2D(64,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPool2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPool2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(4,activation='softmax'))
model.summary()
```

```python
from keras import optimizers
adam=optimizers.Adam(lr=1e-4)
model.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['acc'])
```
